```python
import os
import datetime as dt
import pandas as pd
import shutil
```

```python
onlyfiles = [f for f in os.listdir('/Volumes/GoogleDrive/.shortcut-targets-by-id/1z-OKF0fU2hERBaHoPb6WjyG4FAEBK2vE/Android/2021 files/El Pr (File responses)/Por favor suba su archivo aquí (File responses)') if os.path.isfile(os.path.join('//Volumes/GoogleDrive/.shortcut-targets-by-id/1z-OKF0fU2hERBaHoPb6WjyG4FAEBK2vE/Android/2021 files/El Pr (File responses)/Por favor suba su archivo aquí (File responses)/', f))]
```

```python
data_files = [] 

for file_i in onlyfiles:
    
    path_i = os.path.join('/Volumes/GoogleDrive/.shortcut-targets-by-id/1z-OKF0fU2hERBaHoPb6WjyG4FAEBK2vE/Android/2021 files/El Pr (File responses)/Por favor suba su archivo aquí (File responses)', file_i)
    path_blinded_i = os.path.join('/Volumes/GoogleDrive/.shortcut-targets-by-id/1z-OKF0fU2hERBaHoPb6WjyG4FAEBK2vE/Android/2021 files/El Pr (File responses)/blinded', file_i)
    time = os.path.getctime(path_i)
    shutil.copyfile(path_i, path_blinded_i)
    time_shifted = dt.datetime.fromtimestamp(time+10).strftime('%Y-%m-%d %H:%M:%S')
    time = dt.datetime.fromtimestamp(time).strftime('%Y-%m-%d %H:%M:%S')
    data_files.append([file_i, time, time_shifted])

data_files_df = pd.DataFrame(data_files, columns=['file', 'time', 'time_shifted'])

data_files_df['time'] = pd.to_datetime(data_files_df['time'])
data_files_df['time_shifted'] = pd.to_datetime(data_files_df['time_shifted'])
```

```python
responses_df = pd.read_csv('/Volumes/GoogleDrive/.shortcut-targets-by-id/1z-OKF0fU2hERBaHoPb6WjyG4FAEBK2vE/Android/2021 files/El Pr (File responses)/responses.csv')
responses_df['Timestamp'] = pd.to_datetime(responses_df['Timestamp'], format='%m/%d/%Y %H:%M:%S')
```

```python
Id_from = 1
data_codes = [] 

for index_files, row_files in data_files_df.iterrows():
    Id = Id_from + index_files
    for index_responses, row_responses in responses_df.iterrows():
        if(row_responses['Timestamp']>row_files['time'] and row_responses['Timestamp']<row_files['time_shifted']):
            data_codes.append(["S" + str(Id).zfill(3), row_files['file'], row_responses['Email Address']])

df_codes = pd.DataFrame(data_codes, columns=['id', 'file', 'email'])
df_codes.to_csv('/Volumes/GoogleDrive/.shortcut-targets-by-id/1z-OKF0fU2hERBaHoPb6WjyG4FAEBK2vE/Android/2021 files/El Pr (File responses)/codes.csv', index=False)
```
